```python
import asyncio

from gql import Client, gql
from gql.transport.aiohttp import AIOHTTPTransport
```

```python

players = {}
async def main():

    transport = AIOHTTPTransport(url="http://localhost:8080/query")

    # Using `async with` on the client will start a connection on the transport
    # and provide a `session` variable to execute queries on this connection
    async with Client(
        transport=transport,
        fetch_schema_from_transport=True,
    ) as playerSession:

        # Execute single query
        query = gql(
    """
    {
  players {
    name
  }
}
"""
)
        global players
        players = await playerSession.execute(query)

await main()
```

```python

results = []
count=0
for player in players.get("players"):
    playername = player['name']
    async def main():

        transport = AIOHTTPTransport(url="http://localhost:8080/query")

        # Using `async with` on the client will start a connection on the transport
        # and provide a `session` variable to execute queries on this connection
        async with Client(
            transport=transport,
            fetch_schema_from_transport=True,
        ) as playerSession:

            # Execute single query
            query = gql(
        """
        query projectionResults($name:String) {
    player(input:{name:$name}) {
        name
        games(input: {season:"2021-22"}) {
        projections {
            sportsbook
            targets {
            target
            type
            }
            date
        }
        points
        assists
        rebounds
        steals
        blocks
        turnovers
        three_pointers_made
        date
        }
    }
    }
    """
    )
            global result
            result = await playerSession.execute(query, {"name": playername})
            results.append(result)
    count+=1

    await main()
```

```python
over = 0
under = 0
draw = 0
for result in results:
    # print(result.get("player"))
    for game in result.get('player').get("games"):
        if len(game.get("projections")) > 0:
            for projection in game.get("projections")[0].get('targets'):
                if projection.get('type') == 'Points':
                    if game.get("points") > projection.get('target'):
                        over += 1
                    elif game.get("points") < projection.get('target'):
                        under += 1
                    else:
                        draw += 1
print("POINTS")
total = over + under + draw
print("over:", over, over/total)
print("under:", under, under/total)
print("Draw:", draw, draw/total)
```
